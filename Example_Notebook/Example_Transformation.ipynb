```python
files_dir='/Users/Mikejmnez/Desktop/JHU/Poseidon/ECCO_Data/'

vars_tiled = ['THETA','UVEL', 'VVEL']
ds = []
for var in vars_tiled:
    ds1 = xr.open_dataset(files_dir+var+'_2017_01.nc')
    ds.append(ds1)
ds = xr.merge(ds).drop(['XC', 'YC'])
ds = ds.drop_dims('nv').reset_coords()
ds_grid = xr.open_dataset(files_dir+'ECCO-GRID.nc').reset_coords()
ds = xr.merge([ds, ds_grid])
```

# Create Oceanspy Object

```python
od = ospy.OceanDataset(ds)
metadata={}
cat, entries, url, intake_switch = _find_entries('LLC90',None)
entry=entries[-1]
mtdt = cat[entry].metadata
rename = mtdt.pop('rename', None)
od._ds = od._ds.rename(rename)
metadata = {**metadata, **mtdt}
for var in ['aliases', 'parameters', 'name', 'description', 'projection']:
    val = metadata.pop(var, None)
    if val is not None:
        od = eval('od.set_{}(val)'.format(var))
grid_coords = metadata.pop('grid_coords', None)
if grid_coords is not None:
    od = od.set_grid_coords(**grid_coords)
face_connections = metadata.pop('face_connections', None)
if face_connections is not None:
    od = od.set_face_connections(**face_connections)
vars_mates = [ 'UVEL', 'VVEL', 'dxC', 'dyC', 'dxG', 'dyG', 'hFacW','hFacS', 'rAw', 'rAs']  #
for k in range(int(len(vars_mates) / 2)):
    nk = 2 * k
    od._ds[vars_mates[nk]].attrs['mates'] = vars_mates[nk + 1]
    od._ds[vars_mates[nk+1]].attrs['mates'] = vars_mates[nk]
```

## Surface (relative) vorticity

This requires the creation of a grid object

```python
import xgcm
```

```python
grid = xgcm.Grid(DS, periodic='X')
DepthG=grid.interp(grid.interp(DS['Depth'], axis='X', boundary='extrapolate'), axis='Y', boundary='extrapolate')
u_y = grid.diff(DS.UVEL*DS.dyG, axis='Y', boundary='fill') / DS.rAz
v_x = grid.diff(DS.VVEL*DS.dxG, axis='X', boundary='fill') / DS.rAz
zeta = v_x - u_y
```

```python
DepthG.plot(cmap='Greys_r', add_colorbar=False)
((1.25e4)*zeta.isel(Z=0, time=0).where(DepthG>0)).plot(cmap='seismic', vmin=-.05, vmax=0.05)
plt.show()
```

**Figure**. Snapshot of (scaled) surface relative vorticity.

## Centering the Data at the Arctic Cap (limited faces)

```python
dsnew = LLCtransformation.arctic_centered(od._ds, varlist=['Depth', 'THETA', 'UVEL', 'VVEL','dxC', 'dyC', 'dxG', 'dyG','rAw', 'rAs','rAz'], centered='Pacific')
```

```python
grid = xgcm.Grid(dsnew)
DepthG = grid.interp(grid.interp(dsnew['Depth'], axis='X', boundary='extrapolate'), axis='Y', boundary='extrapolate')
DepthU = grid.interp(dsnew['Depth'], axis='X', boundary='extrapolate')
DepthV = grid.interp(dsnew['Depth'], axis='Y', boundary='extrapolate')

u_y = grid.diff(dsnew.UVEL * dsnew.dyG, axis='Y', boundary='fill') / dsnew.rAz
v_x = grid.diff(dsnew.VVEL * dsnew.dxG, axis='X', boundary='fill') / dsnew.rAz
zeta = v_x - u_y
```

```python
ds = od._ds
Nx = len(ds['X'])
Ny = len(ds['X'])
faces = np.array([2, 5, 6, 7, 10])
tNx = np.arange(0,  3 * Nx + 1, Nx)
tNy = np.arange(0,  3 * Ny + 1, Ny)
chunksX, chunksY = make_chunks(tNx, tNy)
ix = [0,1,1,2,1]
jy = [1,0,1,1,2]
psX = []
psY = []
for i in range(len(ix)):
    psX.append(chunksX[ix[i]])
    psY.append(chunksY[jy[i]])
dsnew = make_array(ds, 3 * Nx,3 *  Ny)
metrics=[ 'dxC', 'dyC', 'dxG', 'dyG', 'hFacW','hFacS', 'rAw', 'rAs']
```

```python
nrot = np.array([5, 6, 7])
rot = np.array([2])
irot = np.array([10])
```

```python
varlist = ['Depth', 'THETA', 'UVEL', 'VVEL', 'dxG', 'dyG', 'dxC', 'dyC', 'rAz']
```

```python
for varName in varlist:
    vName = varName
#     print(varName)
    dims = Dims([dim for dim in ds[varName].dims if dim != 'face'][::-1])
    _shape = tuple([len(dsnew[var]) for var in tuple(dims)[::-1]])
    dsnew[varName] = (dims._vars[::-1],np.nan * np.zeros(_shape))
    dsnew[varName].attrs = ds[varName].attrs
    for k in range(len(faces)):
        fac = 1
        xslice = slice(psX[k][0], psX[k][1])
        yslice = slice(psY[k][0], psY[k][1])
        arg = {dims.X: xslice, dims.Y: yslice}
        data = ds[varName].isel(face=faces[k])
        if faces[k] in nrot:
            dsnew[varName].isel(**arg)[:] = data.values
        else:
            dtr = list(dims)[::-1]
            dtr[-1], dtr[-2] = dtr[-2], dtr[-1]
            if faces[k] in rot: ## face=2
                sort_arg = {'variables': dims.X,'ascending': False}
                if len(dims.X) + len(dims.Y) == 4:
                    if 'mates' in list(ds[varName].attrs):
                        vName = ds[varName].attrs['mates']
                        data = ds[vName].isel(face=faces[k])
                        print(vName, dims.X)
                        if len(dims.Y) == 3 and vName not in metrics:
                            fac=-1
                    _DIMS = [dim for dim in ds[vName].dims if dim != 'face']
                    _dims = Dims(_DIMS[::-1])
                    sort_arg = {'variables': _dims.X,'ascending': False}
            elif faces[k] in irot:
                sort_arg = {'variables': dims.Y,'ascending': False}
                if len(dims.X) + len(dims.Y) == 4:
                    if 'mates' in list(ds[varName].attrs):
                        vName = ds[varName].attrs['mates']
                        data = ds[vName].isel(face=faces[k])
                        if len(dims.X) == 3 and vName not in metrics:
                            print(faces[k], vName)
                            fac= - 1
                    _DIMS = [dim for dim in ds[vName].dims if dim != 'face']
                    _dims = Dims(_DIMS[::-1])
                    sort_arg = {'variables': _dims.Y,'ascending': False}
            data = fac * data.sortby(**sort_arg)
            dsnew[varName].isel(**arg).transpose(*dtr)[:] = data.values
```

```python
DepthU.plot(cmap='Greys_r', add_colorbar=False)
(dsnew['UVEL'].isel(Z=0, time=0).where(DepthU>0)).plot(cmap='seismic', vmin=-.1, vmax=0.1, add_colorbar=False)
plt.show()
```
